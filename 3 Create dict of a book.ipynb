```python
import string
```

```python
def process_file(filename):
    hist = dict()
    fp = open(filename)
    for line in fp:
        print(line)
        process_line(line, hist)
    return hist
```

```python
def process_line(line, hist):
    line = line.replace('-', ' ') 
    for word in line.split():
        word = word.strip(string.punctuation + string.whitespace)
        word = word.lower()
        hist[word] = hist.get(word, 0) + 1
        
```

```python
hist = process_file('C:\\Users\\lifei\\OneDrive\\Desktop\\emma.txt')
```
